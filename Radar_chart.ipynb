```python
import pandas as pd
import math

#create a dataframe for our alternative analysis matrix
AA=pd.DataFrame(columns=['criteria', 'weight', 'ratingA', 'ratingB', 'scoreA', 'scoreB'])
print(AA)
```

```python
#first step: add a list of criteria
AA['criteria']=['risk', 'ROI', 'customerSatisfaction', 'feasibility', 'starategicAlignment']
print(AA)
```

```python
#second step: identify weight
#weight show relative impoetance of each criterion
#they must add up to 1 (or 100 if you use percentages)
AA['weight']=[0.1,0.15,0.3,0.15,0.3]
print(AA)
```

```python
#step 3: rate each alternative across all criteria
#choose a scale: 1-5 or 1-7 or 1-10
#higher number+ an alternative is doing better regarding a given critertion
AA['ratingA']=[4,1,2,1,5]
AA['ratingB']=[3,4,2,3,2]

#step 4: calculate partial scores by multip;ying weight * ratingsr
for index, row in AA.iterrows():
  AA['scoreA'][index]=row['ratingA']*row['weight']
  AA['scoreB'][index]=row['ratingB']*row['weight']

print(AA)
```

```python
totalScoreA=0
totalScoreB=0
for index, row in AA.iterrows():
  totalScoreA+=row['scoreA']
  totalScoreB+=row['scoreB']

print('the total score for A is {:.2f} and for B is {:.2f}'.format(totalScoreA, totalScoreB))
```

```python
import numpy as np
import matplotlib.pyplot as plt

axisLocations = np.linspace(start=0, stop=2 * np.pi, num=len(AA['ratingA']), endpoint=False)
axisLocations = np.concatenate((axisLocations,[axisLocations[0]]))
ratingsA=AA['ratingA']
ratingsB=AA['ratingB']
ratingsACricular=np.concatenate((ratingsA,[ratingsA[0]]))
ratingsBCricular=np.concatenate((ratingsB,[ratingsB[0]]))

plt.figure(figsize=(8, 8))
plt.subplot(polar=True)
plt.plot(axisLocations, ratingsACricular, 'o-', color='red', linewidth=2, label='In-house development')
plt.plot(axisLocations, ratingsBCricular, 'd-', color='blue', linewidth=2, label='Outsourc')

plt.title('Project Options Comparision', size=20)
lines, labels = plt.thetagrids(np.degrees(axisLocations), labels=AA['criteria'])
plt.legend()
plt.show()
```
